```python
df=pd.read_csv('car1.csv')
df.info()
```

```python
def convert_fa_to_en(x):
    if isinstance(x, str):
        return x.translate({ord('۰'): '0', ord('۱'): '1', ord('۲'): '2', ord('۳'): '3', 
                             ord('۴'): '4', ord('۵'): '5', ord('۶'): '6', ord('۷'): '7', 
                             ord('۸'): '8', ord('۹'): '9'})
    else:
        return x

df = df.applymap(convert_fa_to_en)
df.head(3)
```

```python
df.info()
```

```python
df['LifeSpan']=df['LifeSpan'].str.replace('٬','')
df['LifeSpan']=pd.to_numeric(df['LifeSpan'],downcast='signed')
df['InsuranceDeadline'].fillna('0', inplace=True)

# Extract the numerical values from the column of strings
df["YearOfConstruction"] = df["YearOfConstruction"].str.extract("(\d+)").astype(int)
df["InsuranceDeadline"] = df["InsuranceDeadline"].str.extract("(\d+)").astype(int)
```

```python
df.head(3)
```

```python
# Replace 'غیرقابل نمایش' with NaN
df["Price"] = df["Price"].replace("غیرقابل نمایش", '0')

# Remove non-numeric characters from the 'Price' column
df["Price"] = df["Price"].str.replace("[^0-9]", "", regex=True)

# Convert the 'Price' column to numeric data type
df["Price"] = pd.to_numeric(df["Price"], errors="coerce")

# Replace 'NaN' values with a default value
df["Price"] = df["Price"].fillna(0)

# Convert the 'Price' column to integer data type
df["Price"] = df["Price"].astype(np.int64)

df["Price"] 
```

```python
def convert_to_shamsi(df, col_name):
    def convert_to_shamsi(miladi):
        shamsi = miladi - 621 
        return shamsi
    #why 1900? becuase we don't have cars produced before in the dataframe 1900
    df[col_name] = df[col_name].apply(lambda x: convert_to_shamsi(x) if x > 1900 else x)
    return df.head(5)
convert_to_shamsi(df, 'YearOfConstruction')
```

```python
import re
words_to_check = ['جدید','موتور جدید','سال','دوگانه سوز',
         'بنزینی', 'گازوئیل','دوگانه‌سوز شرکتی',
         'دوگانه‌سوز دستی','CNG','-','شخصی','دوگانه‌سوز']

for word in words_to_check:
    mask = df['Brand'].str.contains(word, regex=False)
    mask &= df['Fuel'].isnull()
    if 'بنزینی' in word:
        df.loc[mask, 'Fuel'] = 'بنزینی'
    elif 'دوگانه‌سوز' in word:
        df.loc[mask, 'Fuel'] = 'دوگانه‌سوز'
    elif 'دوگانه سوز' in word:
        df.loc[mask, 'Fuel'] = 'دوگانه‌سوز'
        

for i in words_to_check:
    mask = df['Brand'].str.contains(i, regex=False)
    df.loc[mask, 'Brand'] = df.loc[mask, 'Brand'].str.replace(i,'')
```

```python
def update_gearbox_from_brand(df, words_to_check):
    for word in words_to_check:
        mask = df['Brand'].str.contains(word, regex=False)
        mask &= df['Gearbox'].isnull()
        if 'دنده‌ای' in word:
            df.loc[mask, 'Gearbox'] = 'دنده‌ای'
        elif 'اتوماتیک' in word:
            df.loc[mask, 'Gearbox'] = 'اتوماتیک'
        elif 'دنده ای' in word:
            df.loc[mask, 'Gearbox'] = 'دنده‌ای'
        
    return df
words_to_check = ['دنده‌ای', 'اتوماتیک']
df = update_gearbox_from_brand(df, words_to_check)
```

```python
words_to_check = ['دنده‌ای', 'اتوماتیک', 'دنده ای']
for i in words_to_check:
    mask = df['Brand'].str.contains(i, regex=False)
    df.loc[mask, 'Brand'] = df.loc[mask, 'Brand'].str.replace(i,'')
```

```python
# Calculate the total distance traveled by each car
total_distance = df.groupby(['Brand', 'YearOfConstruction'])['LifeSpan'].sum()

# Calculate the number of years each car has been driven
years_driven = df.groupby(['Brand', 'YearOfConstruction'])['YearOfConstruction'].nunique()

# Calculate the average distance traveled by each car per year
avg_distance_per_year = (total_distance / years_driven).mean()

print(f'The average distance traveled by each car per year is {avg_distance_per_year} kilometers.')
```

```python
# group by Brand and average by Price
grouped = df.groupby('Brand')['Price'].mean().reset_index()

# check if any rows in Gearbox col is null and Price>= 1e9
mask = (df['Gearbox'].isnull()) & (df['Price'] >= 1e9)

# add "اتوماتیک" to Gearbox col where mask is True
df.loc[mask, 'Gearbox'] = 'اتوماتیک'

# join the grouped data with the original DataFrame to get the final result
result = pd.merge(grouped, df[['Brand', 'Gearbox']], on='Brand', how='left').drop_duplicates('Brand')
```

```python
# Create a dictionary of brand to gearbox and fuel mappings
brand_mapping = df.groupby('Brand').agg({'Gearbox': 'first', 'Fuel': 'first'}).to_dict()

# Fill the empty rows in the "Gearbox" and "Fuel" columns based on the corresponding brand
df['Gearbox'] = df.apply(lambda row: brand_mapping['Gearbox'].get(row['Brand'], row['Gearbox']), axis=1)
df['Fuel'] = df.apply(lambda row: brand_mapping['Fuel'].get(row['Brand'], row['Fuel']), axis=1)
```

```python
# filter the dataframe to remove rows where the "Price" column is less than 45e6
df = df[df["Price"] >= 45000000]
```
